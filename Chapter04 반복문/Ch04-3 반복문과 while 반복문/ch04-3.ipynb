#### 시작하기 전에 
***range***는 반복문과 함께 많이 사용되는 자료형. 특정한 횟수 만큼 반복해서 실행하고자 할 때 사용

> 범위

```python
# 범위(range) 자료형 선언(생성)
r = range(10)
print("r", r, "타입:", type(r))

r1 = range(5)         # 0 ~ 5 : 0 1 2 3 4
r2 = range(0,2)       # 0 ~5 : 0 1 2 3 4
r2 = range(0, 5, 1)   # 0 ~5 : 0 1 2 3 4

r2 = range(0, 5, 2)   # 0 ~5 : 0 1 2 3 4
# range(stop) -> range object range(stat, stop step[)]) to stopo
```

> 리스트(list)와 함께 사용하기

```python
list("Python")
list(range(1, 10, 2))

range(1, 11)
range(1, 10 + 1)

# 100 보다 작은 3의 배수 리스트 생성하세요

a = list(range(3, 100, 3))
print(a)
```

```python
# 카운트 다운
import time
for counter in range(10, 0, -1):
    print(counter, end = ' ')
    time.sleep(1)
print("fire!")
```

```python
subjects = ["국어", "영어", "수학", "컴퓨터", "물리"]
for i in range(len(subjects)):
    print(i + 1, "교시", subjects[i])
```

> while 반복문

```python
# 1. for looping   : 특정 횟수 만큼 반복하는 경우(반복 횟수가 정해져 있을 때)
# 2. while looping : 특정 조건에 대하여 반복하는 경우(조건식이 만족할 때까지 반복)
# while bool 표현식:

# 횟수 반복
for i in range(15):
    print(i, end = " ")
    time.sleep(0.5)
print("for 반복문 종료")

print()

# 조건 반복
n = 0
while n < 15:
    print(n, end = " ")
    n += 1
    time.sleep(0.5)
print("while 반복문 종료")
```

```python
import string
string.punctuation

'!"#h$%&e\'(l)*+,-./:;<l=>?@[\\]o^_`{|}~'
# 반복문을 돌려서 숨겨진 단어를 찾으세요

signs = """!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~"""
word = ""
for ch in signs:
    if ch not in string.punctuation:
        word += ch

print("word :", word)
```

> while 반복문: 시간을 기반으로 반복하기

```python
# 세계 표준시(UTC) : 1970.01.01.00:00:00 기준 현재까지의 누적 초

import time
print("UTC : ", time.time())

count = 0
start_time = time.time()
target_time = start_time + 1

while time.time() < target_time:
    count += 1 

end_time = time.time()
print("{}초 동안 {}회 반복했습니다.".format(end_time - start_time, count))
```
