```python
import string
string.punctuation

'!"#h$%&e\'(l)*+,-./:;<l=>?@[\\]o^_`{|}~'
# ë°˜ë³µë¬¸ì„ ëŒë ¤ì„œ ìˆ¨ê²¨ì§„ ë‹¨ì–´ë¥¼ ì°¾ìœ¼ì„¸ìš”

signs = """!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~"""
word = ""
for ch in signs:
    if ch not in string.punctuation:
        word += ch

print("word :", word)
```

```python
# ì£¼ì‚¬ìœ„ ë˜ì§€ê¸°
import random



while True:
    user_key = input("ì•„ë¬´ í‚¤ë‚˜ ì…ë ¤í•˜ì„¸ìš”(ì¢…ë£Œí‚¤:Q/q)>")
    if user_key in ['Q', 'q']:
        break      # ë°˜ë³µë¬¸ ë¹ ì ¸ë‚˜ê°€ê¸°
    else:
        dice = int(random.random()*6) + 1
        print('â— ' * dice)
print("ì£¼ì‚¬ìœ„ ë˜ì§€ê¸°ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤. Good Bye~")
    
```

```python
# continue : ì‹¤í–‰ì„ ê±´ë„ˆë›°ê³  ë°˜ë³µì¡°ê±´ìœ¼ë¡œ ë˜ëŒë¦¬ê¸°

# 369 ê²Œì„
# 3, 6, 9 : 'ğŸ™'
# 36 : 'ğŸ™ğŸ™'

user = input('369ê²Œì„ ìˆ«ì ë²”ìœ„ë¥¼ ì…ë ¥í•˜ì„¸ìš”(ì˜ˆ:1-10) > ')
li = user.split('-')
start = int(li[0])
stop = int(li[1])
# start,2,3,4,5,6,7...stop
for num in range(start, stop + 1):
    count = 0     # ì´ˆê¸°í™”
    for ch in str(num):   # ch = '1230679'
        if ch in ['3', '6', '9']:
            count += 1
    if count == 0:
        continue
    else:
        print('{}: {}'.format(num, 'ğŸ™ ' * count))
```
